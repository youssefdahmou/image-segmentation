```python
import matplotlib.pyplot as plt
from skimage import data
from skimage.filters import threshold_local, threshold_otsu
from skimage import feature
from skimage.color import rgb2gray

# Load an example image (you can replace this with your own image)
image = data.camera()

# Display the image using Matplotlib
plt.imshow(image)
plt.title('Cameraman Image')
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

##  2) -  Segmentation bas´ee sur l’intensit´e (Thresholding)

```python

# 1. Global Thresholding
threshold = 127
image_binarized = image > threshold
plt.imshow(image_binarized, cmap='gray')
plt.show()
```

```python
# 2. Adaptive Thresholding
block_size = 35
offset = 10
threshold_adaptive = threshold_local(image, block_size, offset=offset)
image_binarized_adaptive = image > threshold_adaptive
plt.imshow(image_binarized_adaptive, cmap='gray')
plt.show()
```

```python
# 3. Otsu Thresholding
threshold_otsu_value = threshold_otsu(image)
image_binarized_otsu = image > threshold_otsu_value
plt.imshow(image_binarized_otsu, cmap='gray')
plt.show()
```

```python
# 4. Multi-level Thresholding
threshold1 = 85
threshold2 = 170
mask1 = (image <= threshold1)
mask2 = (image > threshold1) & (image <= threshold2)
mask3 = (image > threshold2)
image_segmented = mask1 + 2 * mask2 + 3 * mask3
plt.imshow(image_segmented, cmap='nipy_spectral')
plt.show()
```

```python
# 5. Hysteresis Thresholding
low_threshold = 30
high_threshold = 100
edges = feature.canny(image, low_threshold=low_threshold, high_threshold=high_threshold)
plt.imshow(edges, cmap='gray')
plt.show()
```

## 3) - Segmentation par croissance de r´egion (Region Growing)

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

def region_growing(image, seed):
    height, width = image.shape
    visited = np.zeros_like(image, dtype=bool)  # Use built-in bool type
    segmented = np.zeros_like(image, dtype=bool)  # Use built-in bool type
    list_pixels = [seed]
    intensity_threshold = 20  # Adjust according to the image

    while len(list_pixels) > 0:
        current_pixel = list_pixels.pop()
        x, y = current_pixel

        if visited[y, x]:
            continue

        visited[y, x] = True
        segmented[y, x] = True

        for dx in [-1, 0, 1]:
            for dy in [-1, 0, 1]:
                nx, ny = x + dx, y + dy
                if 0 <= nx < width and 0 <= ny < height:
                    difference = abs(int(image[y, x]) - int(image[ny, nx]))
                    if not visited[ny, nx] and difference < intensity_threshold:
                        list_pixels.append((nx, ny))

    return segmented

# Load the camera image
image = data.camera()

# Define the seed point
seed_point = (50, 50)  # Choose according to the image

# Perform region growing segmentation
segmented_image = region_growing(image, seed_point)

# Display the segmented image
plt.imshow(segmented_image, cmap='gray')
plt.show()
```

*1) - Le choix du point initial (germe) influence fortement le résultat final de la segmentation par croissance de région. Si le germe est placé dans une région de l'image avec une intensité similaire à celle que vous souhaitez segmenter, la méthode peut réussir. Cependant, un mauvais choix de germe peut entraîner une segmentation incorrecte ou une sous-segmentation.

*2) - Avantages de la segmentation par croissance de région :

        -Elle est simple à comprendre et à mettre en œuvre.
        -Elle peut détecter des régions homogènes dans l'image en se basant sur la similarité des niveaux de gris.
* Inconvénients :

        -Sensible au choix du germe initial.
        -Peut être lente sur des images de grande taille.
        -Moins efficace lorsque les objets à segmenter ont des contours flous ou des textures complexes.

*3)  Pour améliorer la méthode et la rendre plus robuste face aux variations d'intensité dans une région, vous pouvez :

        -Utiliser des critères de similarité plus sophistiqués, comme des modèles statistiques pour la similarité des niveaux de gris.
        -Appliquer un prétraitement pour corriger les variations d'intensité dans l'image.
        -Utiliser des techniques de post-traitement pour éliminer les régions non désirées ou fusionner des régions similaires.
        
*4) Cette méthode est particulièrement utile dans les situations suivantes :

        -Segmenter des objets dont les frontières sont bien définies et qui sont caractérisés par des niveaux de gris homogènes.
        -Identifier des régions d'intérêt dans des images médicales, telles que la segmentation de tumeurs dans des scans.
        -Segmenter des objets simples dans des images de faible complexité, telles que la séparation d'objets sur un fond uniforme.

## 4)- Segmentation par Clustering Objectif :


```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from sklearn.cluster import KMeans

# Load the camera image
image = data.camera()

# Reshape the image into a list of pixels
image_reshaped = image.reshape((-1, 1))  # Convert to a single channel

k = 3  # Number of clusters, adjust as needed
kmeans = KMeans(n_clusters=k)
kmeans.fit(image_reshaped)
clustered = kmeans.cluster_centers_[kmeans.labels_]
segmented_image = clustered.reshape(image.shape)

# Display the segmented image
plt.imshow(segmented_image.astype(np.uint8), cmap='gray')
plt.show()
```

* 1)--  Pour déterminer la valeur optimale de k (nombre de clusters) pour une image donnée, on peut utiliser des techniques telles que la méthode du coude (Elbow method) en traçant la somme des carrés des distances intra-cluster en fonction de k. La valeur de k au point où l'inertie commence à se stabiliser est souvent choisie comme le nombre optimal de clusters.

* 2)--  * Avantages de la segmentation par clustering :

        -Peut identifier automatiquement des structures ou des groupes dans les données.
        -Peut être utilisée pour segmenter des images sans nécessiter une connaissance préalable des objets.
    *Inconvénients :

        -Sensible au nombre de clusters choisi (k).
        -Peut produire des segments de forme régulière, même lorsque les objets sont de forme irrégulière.
        -La performance dépend de l'initialisation des centres, ce qui peut conduire à des résultats suboptimaux.
* 3)--  Pour combiner le clustering basé sur la couleur avec d'autres caractéristiques, on peut utiliser des techniques de segmentation multimodale. Cela implique de combiner les informations de couleur, de texture, de position spatiale, etc. dans un algorithme de segmentation, permettant ainsi une meilleure séparation des objets ayant des caractéristiques diverses.

* 4)--  Le k-means est sensible à l'initialisation des centres, ce qui signifie que différents points de départ peuvent conduire à des résultats de clustering différents. Pour remédier à ce problème, on peut utiliser des initialisations multiples, comme la méthode k-means++, qui réduit la probabilité de convergence vers un minimum local. Il est également courant de lancer l'algorithme de k-means plusieurs fois avec différentes initialisations et de sélectionner le meilleur résultat en fonction de critères tels que l'inertie du cluster.

## 5) Segmentation bas´ee sur la couleur

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

# Load the camera image
original_image = data.camera()

# Function to segment an image based on the HSV color
def segment_by_color_hsv(image, lower_bound, upper_bound):
    # Convert the image to BGR HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

    # Apply a mask based on the specified color bounds
    mask = cv2.inRange(hsv_image, lower_bound, upper_bound)

    # Use the mask to extract the region of interest from the original image
    segmented_image = cv2.bitwise_and(image, image, mask=mask)

    return segmented_image

# Define the color bounds for green hue
lower = np.array([40, 0, 0])
upper = np.array([80, 255, 255])

# Segment the original image using the bounds
segmented_image = segment_by_color_hsv(original_image, lower, upper)

# Display the segmented image
plt.imshow(segmented_image, cmap='gray')
plt.show()
```

*1)--  Le choix du modèle de couleur influence la segmentation car différents modèles représentent les couleurs de manière différente. Certains modèles, comme HSV, sont plus adaptés pour séparer des objets en fonction de leur teinte, saturation et luminosité, tandis que d'autres modèles, comme RGB, se basent sur les composantes rouge, verte et bleue. Le choix dépend de la nature des objets dans l'image et du but de la segmentation.

*2))--  Les avantages de l'utilisation du modèle HSV pour la segmentation basée sur la couleur sont les suivants :

        -Le modèle HSV sépare la luminosité (Value) de la couleur (Hue et Saturation), ce qui facilite la détection de couleurs indépendamment de l'intensité lumineuse.
        -Il est plus intuitif pour spécifier des plages de teintes et de saturations pour la segmentation.
        -Il permet de mieux isoler des objets ayant différentes teintes, même sous différentes conditions d'éclairage.
*4)--  Pour déterminer les limites inférieure et supérieure pour la segmentation d'une couleur spécifique en HSV, vous pouvez utiliser des outils de traitement d'image pour analyser des échantillons d'images représentant la couleur cible. Par exemple, en utilisant un logiciel d'édition d'image, vous pouvez sélectionner un échantillon de la couleur, convertir ce pixel en HSV et noter les valeurs de teinte (Hue) et de saturation (Saturation) qui représentent la couleur. Vous pouvez ensuite définir les limites en ajustant ces valeurs.

*5)--  Pour améliorer la segmentation basée sur la couleur en utilisant d'autres caractéristiques, comme la texture, vous pouvez combiner des méthodes de segmentation basées sur la couleur avec des méthodes de segmentation basées sur la texture. Cela peut être réalisé en utilisant des techniques de traitement d'image avancées, telles que la segmentation multimodale. En combinant des informations de couleur, de texture et d'autres caractéristiques, vous pouvez obtenir une segmentation plus précise et robuste qui tient compte de la variabilité des objets dans l'image.
